```python
import cv2


# Initialize global variables
clicks = []  # To store the (x, y) coordinates of mouse clicks

def mouse_click(event, x, y, flags, param):
    """ Mouse callback function to capture mouse clicks """
    global clicks

    # If the left mouse button was clicked, record the (x, y) coordinates
    if event == cv2.EVENT_LBUTTONDOWN:
        clicks.append((x, y))

        # If this is the second click (defining the second corner of the rectangle), show the bounding box
        if len(clicks) == 2:
            cv2.rectangle(img, clicks[0], clicks[1], (255, 0, 0), 2)
            cv2.imshow('Image', img)
            print(f"Bounding Box: xmin: {min(clicks[0][0], clicks[1][0])}, xmax: {max(clicks[0][0], clicks[1][0])}, ymin: {min(clicks[0][1], clicks[1][1])}, ymax: {max(clicks[0][1], clicks[1][1])}")

def select_bbox(image_path):
    global img
    img = cv2.imread(image_path)
    cv2.imshow('Image', img)
    cv2.setMouseCallback('Image', mouse_click)

    # Wait until the user presses 'q' to quit
    while True:
        if cv2.waitKey(1) & 0xFF == ord('q') or len(clicks) == 2:
            break

    cv2.destroyAllWindows()

# Example usage
image_path = '/content/MITSloanLogo.png'
select_bbox(image_path)
```

```python
import cv2
from google.colab.patches import cv2_imshow

def draw_bbox(image_path, point1, point2):
    """
    Draw a bounding box on an image given two points and display the image.

    Parameters:
    - image_path: Path to the image.
    - point1: A tuple (x1, y1) for the first point.
    - point2: A tuple (x2, y2) for the second point.
    """
    img = cv2.imread(image_path)
    cv2.rectangle(img, point1, point2, (255, 0, 0), 2)
    cv2_imshow(img)
    print(f"Image: {image_path},{min(point1[0], point2[0])},{max(point1[0], point2[0])},{min(point1[1], point2[1])},{max(point1[1], point2[1])}")

# Example usage with manual points input
image_path = '/content/image_16.jpg'
point1 = (50, 15)  # Replace with your selected point 1
point2 = (99, 80)  # Replace with your selected point 2
draw_bbox(image_path, point1, point2)
```
